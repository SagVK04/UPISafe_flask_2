```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score
```

```python
#is_fraud                 1.000000
#amount                   0.397843
#is_international         0.322809
#failed_transactions      0.146129
#transaction_type         0.045977
#device_type              0.007122
input_sample = sample_df[['is_international','failed_transactions','transaction_type','device_type']]
output_sample = sample_df['is_fraud']
```

```python
input_sample.head()
```

```python
!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
#Handling Class Imbalance
sm = SMOTE(random_state=42)
X_res, Y_res = sm.fit_resample(input_sample, output_sample) #Class is balanced for X-values and Y-values by adding new samples(Resampling)

#Split into Train and Test Dataset
X_train,X_test,Y_train,Y_test = train_test_split(X_res,Y_res,test_size=0.1,random_state=42)

#Scaling with Standardize values -> the features
sc = StandardScaler()
train_input_mid = sc.fit_transform(X_train)
test_input_mid = sc.transform(X_test)

#Dimensionality reduction (optional but helps)
pca = PCA(n_components=0.95)  # Only keeps features having 95% of variance
train_input = pca.fit_transform(train_input_mid)
test_input = pca.transform(test_input_mid)
```

```python
knn = KNeighborsClassifier(
    n_neighbors=13,
    weights='distance',
    metric='manhattan',
)
knn.fit(train_input,Y_train)
```

```python
pred = knn.predict(test_input)
pred_1 = knn.predict_proba(test_input)
pred_1
```

```python
fraud_pred = pred_1[0][1]
print(f"Risk Score: {fraud_pred * 100:.2f}%")
```

```python
print(f"Accuracy Score: {accuracy_score(pred,Y_test)*100:.2f}%")
```

```python
print(f"Precision Score: {precision_score(pred,Y_test)*100:.2f}%")
```

```python
max_pred_score = 0
best_k = 0

# Test a meaningful range of k values (e.g., from 1 to 30)
# A common practice is to use odd numbers to avoid tie-breaking issues.

for k in (1,31,2):
    knn_k = KNeighborsClassifier(
        weights='distance',
        metric='manhattan'
    )
    knn_k.fit(train_input, Y_train)
    pred_k = knn_k.predict(test_input)
    current_accuracy = accuracy_score(pred_k, Y_test)
    if current_accuracy > max_pred_score:
        max_pred_score = current_accuracy
        best_k = k

print(f"Optimal K: {best_k}")
print(f"Maximum Accuracy Score: {max_pred_score * 100:.2f}%")
```

```python
import numpy as np
in_tr_input = input('Was transaction International? YES or NO ?') #Was transaction International?
if in_tr_input.lower() == 'yes':
    in_tr = 1
else:
    in_tr = 0

in_fail_input = input('Was transaction failed? YES or NO ?') #Was transaction failed?
if in_fail_input.lower() == 'yes':
    in_fail = 1
else:
    in_fail = 0

in_type_input = input('Online or POS or ATM or Transfer: Which transaction type ?') #Select your transaction type-> Online or POS or ATM or Transfer
if in_type_input.lower() == 'online':
    in_type = 0
elif in_type_input.lower() == 'pos':
    in_type = 1
elif in_type_input.lower() == 'atm':
    in_type = 2
elif in_type_input.lower() == 'transfer':
    in_type = 3

in_dev_input = input('Tablet or Computer or Mobile: Which device ?') #Select your device-> Tablet or Computer or Mobile
if in_dev_input.lower() == 'tablet':
    in_dev = 0
elif in_dev_input.lower() == 'computer':
    in_dev = 1
elif in_dev_input.lower() == 'mobile':
    in_dev = 2
    
prediction_input = np.array([[in_tr, in_fail, in_type, in_dev]])
print(prediction_input)

pred_out = knn.predict(prediction_input)
pred_out_1 = knn.predict_proba(prediction_input)
if pred_out.astype(int) == 1:
    print("Possibly Fraud!")
else:
    print("Possibly Safe!")
print(f"Risk Score: {pred_out_1[0][1]*100:.2f}%")
```
